```python
from pprint import pprint
gkey = ["AIzaSyBC4IyVW7a-mYc-lFFw_qXw94Vo-Rv0PJU"]
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
target_radius = 50000
location_dict = {
    "bar":{"names":[],"coords":[]},
    "bus_station":{"names":[],"coords":[]},
    "car_rental":{"names":[],"coords":[]},
    "embassy":{"names":[],"coords":[]},
    "lodging":{"names":[],"coords":[]},
    "museum":{"names":[],"coords":[]},
    "night_club":{"names":[],"coords":[]},
    "restaurant":{"names":[],"coords":[]},
    "subway_station":{"names":[],"coords":[]},
    "taxi_stand":{"names":[],"coords":[]},
    "train_station":{"names":[],"coords":[]},
    "transit_station":{"names":[],"coords":[]}
}

for coordinate in coord:
    params ={
       "location": coordinate 
    }
    
    for location_type in list(location_dict.keys()):
        params['radius'] = target_radius
        params['type'] = location_type
        params['key']= gkey

        #print(params)

        response = requests.get(base_url, params=params)
        result = response.json()
        results_list = result['results']

        for x in results_list:
            location_dict[location_type]["coords"].append([x["geometry"]["location"]["lat"], x["geometry"]["location"]["lng"]])
            location_dict[location_type]["names"].append(x["name"])
            
pprint(location_dict)
```

```python
# Dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import io
```

```python
lat = airport_df[14]
```

```python
lng = airport_df[15]
```

```python
coord = []

for airport in range(len(lat)):
    coord.append(lat[airport]+" , "+lng[airport])
```

```python
# Testing purposes only
coord = ['33.6558 , -84.4333','40.0724 , 116.583']
```

```python
gkey = ["AIzaSyANl2HVNFDn9ThOju7ELbDumiGMEAQJYds"]
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
target_radius = 1000
location_dict = {
    "bar":{"names":[],"coords":[]},
    "bus_station":{"names":[],"coords":[]},
    "car_rental":{"names":[],"coords":[]},
    "embassy":{"names":[],"coords":[]},
    "lodging":{"names":[],"coords":[]},
    "museum":{"names":[],"coords":[]},
    "night_club":{"names":[],"coords":[]},
    "restaurant":{"names":[],"coords":[]},
    "subway_station":{"names":[],"coords":[]},
    "taxi_stand":{"names":[],"coords":[]},
    "train_station":{"names":[],"coords":[]},
    "transit_station":{"names":[],"coords":[]}
}

for coordinate in coord:
    params ={
       "location": coordinate 
    }
    
    for location_type in list(location_dict.keys()):
        params['radius'] = target_radius
        params['type'] = location_type
        params['key']= gkey

        #print(params)

        response = requests.get(base_url, params=params)
        #if response:
        #    print("OK, RESPONSE")
        result = response.json()
        results_list = result['results']

        for x in results_list:
            location_dict[location_type]["coords"].append([x["geometry"]["location"]["lat"], x["geometry"]["location"]["lng"]])
            location_dict[location_type]["names"].append(x["name"])
            
#pprint(location_dict)
```

```python
# Number of instances per category
print("Number of bars is: " + str(len(location_dict['bar']['names'])))
print("Number of bus stations is: " + str(len(location_dict['bus_station']['names'])))
print("Number of car rentals is: " + str(len(location_dict['car_rental']['names'])))
print("Number of embassies is: " + str(len(location_dict['embassy']['names'])))
print("Number of lodgings is: " + str(len(location_dict['lodging']['names'])))
print("Number of musea is: " + str(len(location_dict['museum']['names'])))
print("Number of night clubs is: " + str(len(location_dict['night_club']['names'])))
print("Number of restaurants is: " + str(len(location_dict['restaurant']['names'])))
print("Number of subway stations is: " + str(len(location_dict['subway_station']['names'])))
print("Number of taxi stands is: " + str(len(location_dict['taxi_stand']['names'])))
print("Number of train stations is: " + str(len(location_dict['train_station']['names'])))
print("Number of transit stations is: " + str(len(location_dict['transit_station']['names'])))
```

```python
# Export
with io.open("all_data_within_1000_meters.json", "w", encoding="utf8") as fp:
    json.dump(location_dict, fp, indent=4, ensure_ascii=False)
```
